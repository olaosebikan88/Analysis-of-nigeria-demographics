# Renaming of the colunms

```python
df = data.rename(columns={'Age Group': 'Age', 
                    'Where do you live?': 'State','Gender' :'Gender', 
                    'Are you Happy with your job?': 'Job_satisfaction', 
                    'Are you married?': 'Marital_status',
                    'Do you have kids?': 'Have_chidren',
                    'What is your steady/constant income range in a month?': 'Monthly_income_range',
                   'Is this your primary source of income?': 'One_income_source',
                   'Are you a Nigerian?': 'Nigerian?', 
                    'Do you pay your tax?': 'Tax_compliance',
                   'What part of your state do you live in?': 'Rural_or_urban',
                   'Do you think it is enough for you and your family?': 'Income_satisfaction',
                   'What are your other sources of income? ': 'Other_income_sources',
                   'What is your largest form of expense': 'Major_expenses',
                   'If there was one thing you really wanted to own, what would that be?': 'Desire',
                   'How much do you think you earn from all other sources of income in a month': 'Other_income',
                   'On an average, how much do you think you spend in a month?': 'Total_monthly_expenses',
                   'Do you think Nigeria can get better?': 'Citizen_sentiment',
                   'In a short sentence, describe your government': 'Comment',
                   'Please fill your email if you want to be notified once this research is completed.': 'Email'})
```

```python
df.columns
```

# Elimination of unwanted colunm

```python
data.drop(columns=['Timestamp'])
```

```python
sns.countplot(x='Monthly_income_range', data=df)
plt.xticks(rotation='vertical')
plt.title('Monthly Income Range')
plt.show()
```

# The highest income range is 21,000-50,000 and the lowest income range is 500,000 and above

```python
sns.countplot(x='Major_expenses', data=df)
plt.xticks(rotation='vertical')
plt.title('Major monthly Expenses')
```

# The highest major expenses is feeding while the lowest is tax, medical bills  and bribery

```python
state_plot = df.groupby(['Monthly_income_range', 'State']).size().reset_index().pivot(columns='Monthly_income_range', index='State', values=0)
fig, (ax1, ax2) = plt.subplots(nrows=2, ncols=1, figsize=(10, 15))
fig.subplots_adjust(hspace=0.4)
state_plot.drop('Lagos', axis=0).plot(kind='bar', stacked=True, ax=ax1)
state_plot.plot(kind='bar', stacked=True, ax=ax2)
```

# Illustration shows more data was sourced from lagos,ogun,oyo,abuja(FTC),kwara and kaduna and less data input from delta,cross river, edo, akwa ibom, nasarawa, niger, benue

```python
fig = plt.figure(figsize=(18, 22))
ax = fig.subplots(nrows=5, ncols=4, squeeze=True)
fig.subplots_adjust(hspace=1.2, wspace=0.3)
groups = df.groupby('State')
states = list(df.State.unique())
i = 0
for axes in ax.flat:
    sns.countplot(x='Monthly_income_range', data=groups.get_group(states[i]), ax=axes)
    plt.sca(axes)
    plt.title('Income Range for {}'.format(states[i]))
    plt.xticks(rotation='vertical')
    i += 1
```

# Illustration of each income ranges for each states

```python
df['Occupation'].fillna('Unknown', inplace=True)
```

```python
occupation_plot = df.groupby(['Monthly_income_range', 'Occupation']).size().reset_index().pivot(columns='Monthly_income_range', index='Occupation', values=0)
occupation_plot.plot(kind='bar', stacked=True, figsize=(6, 10))
```

# Illustrations shows we have more input from software engineers/computer designs, students, engineers,entrepreneurs, students/working, financial services and lowest input from framers

```python
occupation_plot = df.groupby(['Monthly_income_range', 'Tax_compliance']).size().reset_index().pivot(columns='Monthly_income_range', index='Tax_compliance', values=0)
occupation_plot.plot(kind='bar', stacked=True, figsize=(6, 10))
```

# Illstration shows that income range of 500,000 and above are tax invaders while those with smaller income range of 21,000-50,000, 81,000-100,000, 120,000-200,000 complies to tax payment

```python
occupation_plot = df.groupby(['Monthly_income_range', 'Marital_status']).size().reset_index().pivot(columns='Monthly_income_range', index='Marital_status', values=0)
occupation_plot.plot(kind='bar', stacked=True, figsize=(6, 10))
```

# The data set have less input of married than the single

```python
fig = plt.figure(figsize=(18, 24))
ax = fig.subplots(nrows=5, ncols=3)
fig.subplots_adjust(hspace=1.0, wspace=0.3)
groups = df.groupby('Occupation')
industries = list(df.Occupation.unique())
i = 0
axes = ax.flat
for industry in industries:
    sns.countplot(x='Monthly_income_range', data=groups.get_group(industry), ax=axes[i])
    plt.sca(axes[i])
    plt.title('Income Range for those in {} Industry'.format(industry))
    plt.xticks(rotation='vertical')
    i += 1
```
